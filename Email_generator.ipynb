```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://www.linkedin.com/jobs/")
page_data = loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_extract = PromptTemplate.from_template(
    """
    ### SCRAPED TEXT FROM WEBSITE:
        {page_data}
        ### INSTRUCTION:
        The scraped text is from the career's page of a website.
        Your job is to extract the job postings and return them in JSON format containing the 
        following keys: `role`, `experience`, `skills` and `description`.
        Only return the valid JSON.
        ### VALID JSON (NO PREAMBLE):    
        """
)

chain_extract = prompt_extract | llm
res = chain_extract.invoke(input={'page_data':page_data})
print(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser

json_parser = JsonOutputParser()
json_res = json_parser.parse(res.content)
json_res
```

```python
import chromadb
import uuid

# client = chromadb.PersistentClient()
# collection = client.create_collection(name="portfolio")

import onnxruntime as ort

# Save the original InferenceSession class
OriginalInferenceSession = ort.InferenceSession

# Create a patched version that forces the CPU provider
class PatchedInferenceSession(OriginalInferenceSession):
    def __init__(self, model, sess_options=None, providers=None, provider_options=None):
        # Force providers to CPUExecutionProvider
        super().__init__(model, sess_options, providers=['CPUExecutionProvider'], provider_options=provider_options)

# Monkey-patch onnxruntime.InferenceSession with our patched version
ort.InferenceSession = PatchedInferenceSession

# Now import and use the embedding function
from chromadb.utils.embedding_functions import ONNXMiniLM_L6_V2

# Instantiate the embedding function (it will create its InferenceSession with the CPU provider)
embedding_function = ONNXMiniLM_L6_V2()

# Create your ChromaDB client and collection using the custom embedding function
# import chromadb
# client = chromadb.Client()
# collection = client.create_collection(
#     name="my_collection",
#     embedding_function=embedding_function
# )


import chromadb
import uuid

client = chromadb.PersistentClient('vectorstore')
collection = client.get_or_create_collection(name="portfolio")

if not collection.count():
    for _, row in df.iterrows():
        collection.add(documents=row["Techstack"],
                       metadatas={"links": row["Links"]},
                       ids=[str(uuid.uuid4())])
```
